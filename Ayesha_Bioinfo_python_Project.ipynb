```python
# Correlation Matrix
# Convert 'diagnosis' column to numerical representation before calculating correlation
data_1['diagnosis_numeric'] = data_1['diagnosis'].map({'M': 1, 'B': 0})
# Create a new column with numerical values
corr_matrix = data_1.drop(columns=['diagnosis']).corr()
# Exclude the original 'diagnosis' column from correlation calculation

plt.figure(figsize=(15,10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Cancer Data')
plt.show()
```

```python
# Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='diagnosis', y='radius_mean', data=data_1)
plt.title('Boxplot of Radius Mean by Diagnosis')
plt.show()

# Violinplot
plt.figure(figsize=(10, 6))
sns.violinplot(x='diagnosis', y='texture_mean', data=data_1)
plt.title('Violinplot of Texture Mean by Diagnosis')
plt.show()

# Heatmap of a subset of features
subset_features = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean']
corr_matrix_subset = data_1[subset_features].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix_subset, annot=True, cmap='viridis')  # Different colormap
plt.title('Correlation Matrix of Subset Features')
plt.show()
```

Comparision of distrubition between malignant and benign tumours

```python
#Create separate DataFrames for malignant and benign tumors
malignant_tumors = data_1[data_1['diagnosis'] == 'M']
benign_tumors = data_1[data_1['diagnosis'] == 'B']

# Comparison of Mean Values
mean_comparison = pd.DataFrame({'Feature': data_1.columns[1:],
                                 'Malignant Mean': malignant_tumors.iloc[:, 1:].mean(),
                                 'Benign Mean': benign_tumors.iloc[:, 1:].mean(),
                                 'Difference': malignant_tumors.iloc[:, 1:].mean() - benign_tumors.iloc[:, 1:].mean()})

print(mean_comparison)


# Plotting Histograms for Comparison
features_to_compare = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean']

plt.figure(figsize=(15, 10))
for i, feature in enumerate(features_to_compare):
  plt.subplot(2, 3, i + 1)
  sns.histplot(malignant_tumors[feature], kde=True, color='red', label='Malignant', alpha=0.5)
  sns.histplot(benign_tumors[feature], kde=True, color='blue', label='Benign', alpha=0.5)
  plt.title(f'Distribution of {feature} (Malignant vs. Benign)')
  plt.legend()
plt.tight_layout()
plt.show()

# Box plots for comparison
plt.figure(figsize=(15,10))

for i, col in enumerate(data_1.columns[1:6]):
    plt.subplot(2,3, i+1)
    sns.boxplot(x='diagnosis', y=col, data=data_1)
    plt.title(f"Box Plot of {col} vs Diagnosis")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

def evaluate_model(model_name, y_true, y_pred, model):
    """Evaluates a classification model and prints performance metrics."""
    print(f"\n--- {model_name} Performance ---")
    print(classification_report(y_true, y_pred))
    print(f"Accuracy: {accuracy_score(y_true, y_pred)}")
    cm = confusion_matrix(y_true, y_pred)
    sns.heatmap(cm, annot=True, fmt="d")
    plt.title(f"Confusion Matrix for {model_name}")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()

    # Calculate ROC AUC score
    y_prob = model.predict_proba(X_test_selected)[:, 1]
    roc_auc = roc_auc_score(y_true.map({'B': 0, 'M': 1}), y_prob)
    print(f"ROC AUC Score: {roc_auc}")

    # Plot ROC Curve
    fpr, tpr, thresholds = roc_curve(y_true.map({'B': 0, 'M': 1}), y_prob)
    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend(loc='lower right')
    plt.show()
```
